```python
#1.
#First, we will import the required modules
from bokeh.plotting import figure as fig
from bokeh.plotting import output_notebook as ON
from bokeh.plotting import show
import bokeh.io

# output to the notebook
ON()

# creating figure
plot1 = fig(plot_width = 500, plot_height = 500, title = "Scatter Markers")

# adding a circle renderer with size, color and alpha
plot1.circle([1, 2, 3, 4, 5, 6], [2, 1, 6, 8, 0],
         size = 12, color = "green", alpha = 1)

# show the results
show(plot1)
```

```python
#2.
"""In Bokeh, glyphs are the visual markers or shapes used to represent data points on a plot. They are the building blocks of Bokeh plots, and they allow you to create a wide
   variety of visualizations, such as scatter plots, line charts, bar charts, and more.


   To add glyphs to a Bokeh plot, you use the figure function to create a plot and then call the appropriate glyph
   function to add the desired visual representation of your data."""

#Eaxample
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Prepare the data
x_values = [1, 2, 3, 4, 5]
y_values = [5, 3, 4, 2, 1]

# Create a new plot with specific title and axes labels
p = figure(title='Scatter Plot Example', x_axis_label='X-axis', y_axis_label='Y-axis')

# Add circles as glyphs to the plot
p.circle(x=x_values, y=y_values, size=10, color='blue', alpha=0.8)

# Configure the output mode (e.g., show the plot in Jupyter Notebook)
output_notebook()

# Show the plot
show(p)
```
